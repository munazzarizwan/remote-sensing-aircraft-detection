```python
import os
from PIL import Image

# List of image folders to process
folders = [
    r'C:\Users\dj\Downloads\Michal planes.v1i.yolov8\train\images',
    r'C:\Users\dj\Downloads\Michal planes.v1i.yolov8\valid\images',
    r'C:\Users\dj\Downloads\Michal planes.v1i.yolov8\test\images'
]

# Loop over each folder
for folder_path in folders:
    for filename in os.listdir(folder_path):
        if filename.endswith('.png'):
            img_path = os.path.join(folder_path, filename)
            img = Image.open(img_path).convert('RGB')

            new_filename = filename.replace('.png', '.jpg')
            new_img_path = os.path.join(folder_path, new_filename)

            img.save(new_img_path, 'JPEG')
            os.remove(img_path)  # delete the original PNG

    print(f"✅ Converted all PNGs to JPGs in: {folder_path}")

```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # or yolov8s.pt

results = model.train(
    data=r'C:\Users\dj\Downloads\Michal planes.v1i.yolov8\data.yaml',
    epochs=50,   # Increase from 20 to 50
    imgsz=640
)
```
